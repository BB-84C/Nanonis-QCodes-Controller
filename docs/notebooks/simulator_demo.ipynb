# Nanonis STM Simulator Demo

This notebook demonstrates read-only monitoring and guarded-write planning against STM Simulator.

```python
from qcodes.station import Station

from nanonis_qcodes_controller.qcodes_driver import QcodesNanonisSTM

station = Station()
nanonis = QcodesNanonisSTM("nanonis", auto_connect=True)
station.add_component(nanonis)
nanonis.client_health()
```

```python
snapshot = {
    "bias_v": nanonis.bias_v(),
    "current_a": nanonis.current_a(),
    "zctrl_setpoint_a": nanonis.zctrl_setpoint_a(),
    "scan_status_code": nanonis.scan_status_code(),
}
snapshot
```

```python
# Plan a guarded write (default policy may block live writes).
plan = nanonis.plan_bias_v_set(1.8)
plan
```

```python
# Optional: inspect write audit and close.
nanonis.guarded_write_audit_log()[-5:]
nanonis.close()
```
